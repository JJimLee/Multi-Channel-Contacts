# Multi-Channel Contacts Problem 

```python
# import package
import json
from collections import defaultdict
import pandas as pd
```

```python
# read json file
# Or "/content/drive/MyDrive/shopee21/contacts.json" on google drive
with open('/Users/leo/Desktop/contacts.json', mode = 'r', encoding = 'UTF-8') as file:
    data = json.load(file)
```

```python
# check length of datasets
data_length = len(data) 
```

```python
# convert json to dataframe
record = dict()
record['Id'] = [data[0]['Id']]
record['Email'] = [data[0]['Email']]
record['Phone'] = [data[0]['Phone']]
record['Contacts'] = [data[0]['Contacts']]
record['OrderId'] = [data[0]['OrderId']]
for i in range(1, data_length):
    record['Id'].append(data[i]['Id'])
    record['Email'].append(data[i]['Email'])
    record['Phone'].append(data[i]['Phone'])
    record['Contacts'].append(data[i]['Contacts'])
    record['OrderId'].append(data[i]['OrderId'])
df = pd.DataFrame.from_dict(record)
```

```python
# check first 5 rows of dataset
df.head()
```

```python
# create dictionary to store id and the corresponding contacts
visit = {i: [set([i]), 0] for i in range(500000)}
```

```python
# merge data base on email
find_email = df[['Id', 'Email']][df['Email'] != ''].sort_values(by = 'Email')

for i in range(len(find_email) - 1):
    if find_email.Email.iloc[i] == find_email.Email.iloc[i + 1]:
        visit[find_email.Id.iloc[i]][0].add(find_email.Id.iloc[i + 1])
        visit[find_email.Id.iloc[i + 1]][0].add(find_email.Id.iloc[i])
```

```python
# merge data base on phone
find_phone = df[['Id', 'Phone']][df['Phone'] != ''].sort_values(by = 'Phone')

for i in range(len(find_phone) - 1):
    if find_phone.Phone.iloc[i] == find_phone.Phone.iloc[i + 1]:
        visit[find_phone.Id.iloc[i]][0].add(find_phone.Id.iloc[i + 1])
        visit[find_phone.Id.iloc[i + 1]][0].add(find_phone.Id.iloc[i])
```

```python
# merge data base on orderId
find_orderId = df[['Id', 'OrderId']][df['OrderId'] != ''].sort_values(by = 'OrderId')

for i in range(len(find_orderId) - 1):
    if find_orderId.OrderId.iloc[i] == find_orderId.OrderId.iloc[i + 1]:
        visit[find_orderId.Id.iloc[i]][0].add(find_orderId.Id.iloc[i + 1])
        visit[find_orderId.Id.iloc[i + 1]][0].add(find_orderId.Id.iloc[i])
```

```python
# calculate total contacts (this will take some time to iterate 500,000 rows)
for _, value in visit.items():
    print(_)
    id_list, contacts = value[0], value[1]
    for i in id_list:
        contacts += int(df.Contacts[df['Id'] == i])
    visit[_][1] = contacts
```
